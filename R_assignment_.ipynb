**QUESTION 1 - Create a file that contains 1000 lines of random strings.**

**Question 3: Create a random dataset of 500 rows and 10 columns.**

```r
num_rows <- 500
num_cols <- 10
random_data <- matrix(data = NA, nrow = num_rows, ncol = num_cols)
random_data[, 1:4] <- runif(n = num_rows * 4, min = -10, max = 10)
random_data[, 5:8] <- runif(n = num_rows * 4, min = 10, max = 20)
random_data[, 9:10] <- runif(n = num_rows * 2, min = -100, max = 100)
random_df <- as.data.frame(random_data)
print(random_df)

```

**(i) K-means clustering**

```r
# Apply K-Means clustering and determine the optimal number of clusters
max_clusters <- 10  # Maximum number of clusters to explore
wcss <- numeric(max_clusters)  # Within-cluster sum of squares
for (k in 1:max_clusters) {
  kmeans_model <- kmeans(random_data, centers = k, nstart = 10)  # Run K-Means with 'k' clusters
  wcss[k] <- kmeans_model$tot.withinss  # Store the within-cluster sum of squares
}

# Plot the distance metric graph (WCSS vs. number of clusters)
plot(1:max_clusters, wcss, type = "b", pch = 19, frame = FALSE,
     xlab = "Number of Clusters", ylab = "Within-Cluster Sum of Squares",
     main = "K-Means Clustering - Optimal Number of Clusters")

# Find the optimal number of clusters using the "elbow" method
diff_wcss <- diff(wcss)
elbow_index <- which(diff_wcss <= mean(diff_wcss))

# Add a vertical line at the optimal number of clusters
abline(v = elbow_index, col = "red")
```

**(ii) Hierarchical clustering**

```r
dissimilarity_matrix <- dist(random_data)
hc <- hclust(dissimilarity_matrix, method = "complete")
wss <- (hc$height ^ 2)
plot(hc, hang = -1, main = "Hierarchical Clustering - Optimal Number of Clusters")
diff_wss <- diff(wss)
elbow_index <- which(diff_wss <= mean(diff_wss))
abline(h = wss[elbow_index], col = "red")

```

**Question 5 : Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].**

```r
dataset <- matrix(runif(500 * 5, min = 5, max = 10), nrow = 500)
print(dataset)
```

**(i)Perform t-Test on each column.**

```r
null_value <- 7
t_test_results <- apply(dataset, 2, function(x) t.test(x, mu = null_value))
print(t_test_results)
```

**(ii)Perform Wilcoxon Signed Rank Test on each column**

```r
wilcoxon_test_results <- apply(dataset, 2, function(x) wilcox.test(x, mu = null_value))
print(wilcoxon_test_results)
```

**(iii)Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4**

```r
column_3 <- dataset[, 3]
column_4 <- dataset[, 4]
t_test_result <- t.test(column_3, column_4)
wilcoxon_test_result <- wilcox.test(column_3, column_4)
print(t_test_result)
print(wilcoxon_test_result)
```
